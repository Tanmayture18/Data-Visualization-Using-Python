### Importing pandas library

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df1=df[['zin','indus']]
```

```python
df1.head()
```

### Line plots

```python
df1.plot(kind='line')
```

### Histogram

```python
df['zin'].plot(kind='hist')
plt.xlabel('zin')
plt.title("Histogram")
plt.show()
```

### Area plot

```python
df.plot(kind='area')
```

### Boxplots

```python
df.plot(kind='box')
plt.title('Boxplots')
plt.show()
```

### Scatter plots

```python
df.plot(kind='scatter',x='zin',y='medv')
```

### Heatmap

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(df.corr(),annot=True,cmap='BuPu')
plt.show()
```

### Regression plot

```python
sns.regplot(x='rm',y='medv',data=df,color='green',marker='*')
```
